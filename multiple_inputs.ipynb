```python
from typing import TypedDict, Dict, List
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    values : List[int]
    name :str 
    results : str
```

```python
def process_values(state: AgentState)-> AgentState:
    """this funtion handels multiple inputs"""

    state['results'] = f"Hello {state['name']}, the sum of your values is {sum(state['values'])}"
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("processing", process_values)
graph.set_entry_point("processing")
graph.set_finish_point("processing")
app = graph.compile()
```

```python
from IPython.display import display, Image
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
results = app.invoke({"values": [1,2,3,4,5], "name": "AK"})
```
